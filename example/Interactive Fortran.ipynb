# Interactive Fortran

```python
from numpy import arange
from matplotlib.pyplot import plot
from IPython.html.widgets import interact
%matplotlib inline
```

```python
%load_ext fortranmagic
```

```python
%%fortran
subroutine f1(x, z)
    real, intent(in) :: x
    real, intent(out) :: z

    z = sin(x)

end subroutine f1
```

```python
@interact
def ifort(w=5):
    xx = arange(0,10,0.1)
    yy = [f1(w*x) for x in xx]
    plot(xx,yy)
```
